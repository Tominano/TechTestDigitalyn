```python
#import the packages what we need
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../solar_mesh.csv',index_col=0) 
# We can simply read the file with pandas, fortunately is in the manageable form,
# we have to set the first collumn to index collumn with "index_col=0".
# We can set the new index for the plot as df = df.set_index('bins_lat') but the first line is more formal. 
```

```python
#df # Just check the dataset, how it looks.
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(df, cmap='viridis') # Here is the Heatmap plot
plt.title('Weights in Germany\n based on binned cordinates', fontsize = 20) # title with fontsize 20
plt.xlabel('Latitudes', fontsize = 15) # x-axis label with fontsize 15
plt.ylabel('Longitudes', fontsize = 15) # y-axis label with fontsize 15

#plt.savefig('weights_origin.jpg', dpi=600, format='jpg') Uncmmand this if you would like to save the fig.
```

```python
column_maxes = df.max()
df_max = column_maxes.max()
normalized_df = df / df_max
```

```python
fig, ax = plt.subplots(figsize=(15,15))
ax = sns.heatmap(normalized_df, cmap = 'viridis') # Here is the Heatmap plot
plt.title('Normalized data of weights in Germany\n based on binned cordinates', fontsize = 20) # title with fontsize 20
plt.xlabel('Latitudes', fontsize = 15) # x-axis label with fontsize 15
plt.ylabel('Longitudes', fontsize = 15) # y-axis label with fontsize 15
#plt.savefig('weights_norm.jpg', dpi=600, format='jpg') Uncmmand this if you would like to save the fig.
```
